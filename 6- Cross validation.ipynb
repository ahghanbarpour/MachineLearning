```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/USER/Downloads/Advertising.csv")
df.head()
```

```python
X = df.drop('sales', axis=1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=100)

from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
```

```python
scores
```

```python
abs(scores.mean())
```

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=1)

from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X_train, y_train, scoring='neg_mean_squared_error', cv=5)

abs(scores.mean())
```

```python
model.fit(X_train, y_train)
y_final = model.predict(X_test)

from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_final)
```

```python
from sklearn.model_selection import cross_validate
model = Ridge(alpha=1)

scores = cross_validate(model, X_train, y_train, scoring=['neg_mean_squared_error', 'neg_mean_absolute_error'], cv=15)
scores = pd.DataFrame(scores)
scores
```

```python
scores.mean()
```

```python
model.fit(X_train, y_train)

y_final = model.predict(X_test)
mean_squared_error(y_test, y_final)
```

```python
from sklearn.linear_model import ElasticNet
base_elastic_net_model = ElasticNet()
param_grid = {'alpha':[0.1, 1, 5, 50, 100], 'l1_ratio':[0.1, 0.5, 0.7, 0.95, 0.99, 1]}
```

```python
from sklearn.model_selection import GridSearchCV
grid_model = GridSearchCV(estimator=base_elastic_net_model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5, verbose=0)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_final = grid_model.predict(X_test)
mean_squared_error(y_test, y_final)
```

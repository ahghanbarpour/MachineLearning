```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/USER/Downloads/Advertising.csv')
df.head()
```

```python
X = df.drop('sales', axis=1)
y = df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_converter = PolynomialFeatures(degree=3, include_bias=False)
```

```python
poly_features = poly_converter.fit_transform(X)
X.shape
```

```python
poly_features.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
```

Standardization

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

Standardized Data

```python
X_train
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train, y_train)
```

```python
test_pred = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

MAE = mean_absolute_error(y_test, test_pred)
MSE = mean_squared_error(y_test, test_pred)
RMSE = np.sqrt(MSE)
```

```python
print("MAE: ", MAE)
print("MSE: ", MSE)
print("RMSE: ", RMSE)
```

```python
from sklearn.linear_model import RidgeCV

ridgecv_model = RidgeCV(alphas=(0.1, 1.0, 10.0))
```

```python
ridgecv_model.fit(X_train, y_train)
```

```python
ridgecv_model.alpha_
```

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=0.1)
```

```python
lasso_model.fit(X_train, y_train)
```

```python
test_pred_lasso = lasso_model.predict(X_test)
```

```python
MAE_lasso = mean_absolute_error(y_test, test_pred_lasso)
MSE_lasso = mean_squared_error(y_test, test_pred_lasso)
RMSE_lasso = np.sqrt(MSE_lasso)
```

```python
print("MAE: ", MAE_lasso)
print("MSE: ", MSE_lasso)
print("RMSE: ", RMSE_lasso)
```

```python
from sklearn.linear_model import LassoCV

lassocv_model = LassoCV(eps=0.001, n_alphas=100, cv=5, max_iter=1000000)
```

```python
lassocv_model.fit(X_train, y_train)
```

```python
lassocv_model.alpha_
```

```python
test_pred_lassocv = lassocv_model.predict(X_test)
```

```python
MAE_lassocv = mean_absolute_error(y_test, test_pred_lassocv)
MSE_lassocv = mean_squared_error(y_test, test_pred_lassocv)
RMSE_lassocv = np.sqrt(MSE_lassocv)
```

```python
print("MAE: ", MAE_lassocv)
print("MSE: ", MSE_lassocv)
print("RMSE: ", RMSE_lassocv)
```
